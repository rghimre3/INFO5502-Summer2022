```python
# import necessary packages
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
import ast
import datetime
```

To extract tweets we have created list of hash tags and locations related to Omicron Vaccination & the main cities (capital) in United States.
We will extract the tweets based on keywords & locations provided.

Read the csv file that contains locations for scraping tweets.

```python
df = pd.read_csv('/content/drive/MyDrive/Lilly Paru/locations.csv',sep = ',')
```

#### Query by Text Search

The code below will scrape for 1000 tweets since February 24th, 2022 for each keyword/hashtags by a text search then provide a CSV file with Pandas.

```python
location_list = df.locations.tolist()
hash_tags_list = ['#Omicron', '#Omicron','#VaccinesWork','#COVID19Vaccination','#vaccination','#Vaccination',
                  '#OmicronVaccination','#CovidVaccination','#CovidVaccines','covid-19 booster',
                  '#covid-19 booster','Omicron Vaccination','Omicron','Vaccination']
```

```python
# Creating list to append tweet data to
tweets_list = []

for location in location_list:
    
    for hash_tag in hash_tags_list:
    
        # Setting variables to be used below
        maxTweets = 500
        df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
        '{} near:"{}" within:15km since:2022-03-01 until:2023-03-31'.format(hash_tag, location)).get_items(), maxTweets))
        tweets_list.append(df)
```

```python
# Concatenate list of tweets dataframes
df_concat = pd.concat(tweets_list)
```

```python
# Drop the missing values in place column
df_concat = df_concat.dropna(subset=['place'])
df_concat = df_concat.reset_index(drop=True)
```

```python
# Extract user details from user column
df_concat['GeoCityProvince'] =  df_concat['place'].apply(lambda x: ast.literal_eval(x)['name'])
df_concat['GeoCountry'] = df_concat['place'].apply(lambda x: ast.literal_eval(x)['country'])
df_concat['AuthorID'] =  df_concat['user'].apply(lambda x: eval(x)['id'])
df_concat['UserName'] =  df_concat['user'].apply(lambda x: eval(x)['username'])
df_concat['CreatedAccountAt'] =  df_concat['user'].apply(lambda x: eval(x)['created'])
df_concat['FollowerCount'] =  df_concat['user'].apply(lambda x: eval(x)['followersCount'])
df_concat['FollowingCount'] = df_concat['user'].apply(lambda x: eval(x)['friendsCount'])
df_concat['TweetCount'] =  df_concat['user'].apply(lambda x: eval(x)['statusesCount'])
```

```python
# Renaming the columns
df_rename = df_concat.rename(columns={'date':'CreateDate', 'rawContent':'TweetText', 'id':'TweetID','retweetCount':'RetweetCount',
                         'replyCount':'ReplyCount', 'likeCount':'LikeCount', 'hashtags':'Hashtags'})

df_rename.head()
```

```python
df_rename.columns
```

```python
# Remove unwanted columns

df_drop = df_rename.drop(['Unnamed: 0', 'url','user','renderedContent','quoteCount','conversationId','source','sourceUrl',
                          'sourceLabel','links','media','retweetedTweet','quotedTweet','inReplyToTweetId',
                         'inReplyToUser','mentionedUsers','coordinates','cashtags','card','place','viewCount','vibe'],axis=1)
```

```python
df_drop.head()
```

#### Filter tweets for English language

```python
# Let's find how many tweets are in English
df_en = df_drop[df_drop['lang']=='en']
df_en = df_en.reset_index(drop=True)
df_en.head()
```
